```python
from matplotlib import pyplot as plt
import matplotlib.dates as mdates


def simulate_asset_management(annual_investment: int, annual_interest_rate: float, 
        operation_years: int, start_year: int) -> None:

    amount = 0
    amount_without_investment = 0
    amounts = []
    amounts_without_investment = []
    years = []
    end_year = start_year + operation_years

    for i in range(start_year, end_year):
        amount = (amount + annual_investment) * annual_interest_rate
        amounts.append(amount)
        amount_without_investment += annual_investment
        amounts_without_investment.append(amount_without_investment)
        years.append(i)

    plt.plot(years, amounts)
    plt.plot(years, amounts_without_investment)
    plt.grid()
    plt.ticklabel_format(style='plain',axis='y')
    plt.show()


"""
年間投資額: 120000
    毎月: 10000
インデックスファンドの年利: 1.05
運用年数: 20
開始西暦: 2022
"""
simulate_asset_management(annual_investment=120000, annual_interest_rate=1.05, 
        operation_years = 20, start_year = 2022)

"""
年間投資額: 1216000
    積立NISA: 年額 400000
    iDeCo: 年額 816000(68000 * 12)
インデックスファンドの年利: 1.05
運用年数: 20
開始西暦: 2022
"""
simulate_asset_management(annual_investment=1216000, annual_interest_rate=1.05, 
        operation_years = 20, start_year = 2022)
```
